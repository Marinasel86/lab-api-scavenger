```python
# Challenge 1: Fork Languages.
```

```python
import requests
import json
```

```python
url = "https://api.github.com/users/Marinasel86/repos"
```

```python
response = requests.get(url)
```

```python
response
```

```python
print(response.text)
```

```python
repositories = response.json()
```

```python
repositories
```

```python
languages = []
```

```python
for repository in repositories:
    language = repository.get('language')
```

```python
if language and language not in languages:
            languages.append(language)
```

```python
print(languages)
```

```python
# Challenges 2: Count commits
```

```python
from datetime import datetime, timedelta
```

```python
three_weeks_ago = datetime.now() - timedelta(weeks=3)
```

```python
since_date = three_weeks_ago.isoformat()
```

```python
url2 = "https://api.github.com/users/Marinasel86/repos/commits?since={since_date}"
```

```python
response = requests.get(url2)
```

```python
response
```

```python
print(response.text)
```

```python
commits = json.loads(response.text)
```

```python
num_commits = len(commits)
```

```python
print(f'Number of commits in the past 3 weeks: {num_commits}')
```
